# Basic UI

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle

# Function to predict heart attack risk and message
def predict_heart_attack(age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal):
    features = [[age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal]]
    prediction = model.predict(features)
    if np.any(prediction > 5):



        return prediction, "You have a high risk of heart attack. Please consult a doctor immediately."
    else:
        return prediction, "Your risk of heart attack is low. No immediate concern."

# Get user inputs
age = float(input("Enter age: "))
sex = float(input("Enter sex (0 = female, 1 = male): "))
cp = float(input("Enter cp (0 = typical angina, 1 = atypical angina, 2 = non-anginal pain, 3 = asymptomatic): "))
trestbps = float(input("Enter trestbps (resting blood pressure): "))
chol = float(input("Enter chol (serum cholesterol): "))
fbs = float(input("Enter fbs (fasting blood sugar > 120 mg/dl, 1 = true; 0 = false): "))
restecg = float(input("Enter restecg (resting electrocardiographic results): "))
thalach = float(input("Enter thalach (maximum heart rate achieved): "))
exang = float(input("Enter exang (exercise induced angina, 1 = yes; 0 = no): "))
oldpeak = float(input("Enter oldpeak (ST depression induced by exercise relative to rest): "))
slope = float(input("Enter slope (the slope of the peak exercise ST segment): "))
ca = float(input("Enter ca (number of major vessels colored by fluoroscopy): "))
thal = float(input("Enter thal (3 = normal; 6 = fixed defect; 7 = reversible defect): "))

# Specify the path to the model file
model_path = "/Users/shikarichacha/Desktop/PROJECT /heart_attack_model.pkl"

# Load the model
with open(model_path, 'rb') as file:
    model = pickle.load(file)

# Predict
prediction, message = predict_heart_attack(age, sex, cp, trestbps, chol, fbs, restecg, thalach, exang, oldpeak, slope, ca, thal)

# Create plots
fig, axs = plt.subplots(3)
fig.suptitle('Your Health Stats')
axs[0].hist([age], bins=10, color='blue', alpha=0.7)
axs[0].set_title('Age Distribution')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')

axs[1].hist([chol], bins=10, color='green', alpha=0.7)
axs[1].set_title('Cholesterol Distribution')
axs[1].set_xlabel('Cholesterol')
axs[1].set_ylabel('Frequency')

axs[2].hist([thalach], bins=10, color='red', alpha=0.7)
axs[2].set_title('Heart Rate Distribution')
axs[2].set_xlabel('Heart Rate')
axs[2].set_ylabel('Frequency')

# Show plots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Display prediction and message
print("Prediction:", prediction)
print("Message:", message)
```
